```python
Question 3: You have a pandas DataFrame df with a column named 'values'. Write a Python function that iterates over the dataframe and calculate the
sum of the first three values in the 'values' column. The  function should print the sum to the console.

For example, if the 'values' column of df contains the values[10,20,30,40,50], your function should calculate and print the sum of firsst three values,
which is 60.

Answer: 
# Python function for iterates over the data frame is:
# iterating over rows using iterrows() function

for i, j in df.iterrows():
    print(i,j)
    print()
    
    
# Pyhton function to calculate the sum of the first three values in the 'values' column.
To sum pandas DataFrame columns(given selected multiple columns) using either sum(), iloc[], eval() and loc[] functions. Among these pandas DataFrame.sum() 
functions returns the sum of the values for the requested axis, In order to calculate the sum of the columns use axis = 1

# Select 1 to 3 columns to sum
df['Sum'] = df.iloc[:,1:3].sum(axis = 1)
```

```python
Question 8: You have a Pandas DatFrame df with 'A', 'B', 'C'. Write a Python functions that selects all rows where the value in column 'A' is greater than 5
and value in column 'B' is less than 10. Then function should return a new DataFrame that contains only the selected rows.
For example, if df contains the following values:
    A                 B                C
0   3                5                 1
1   8                2                 7
2   6                9                 4
3   2                3                 5
4   9                1                 2


Answer: 
# the function to create dataframe of the given situation

import pandas as pd
data = {'A': [ 3,8,6,2,9],
        'B': [5,2,9,3,1],
        'c': [1,7,4,5,2]}
df = pd.DataFrame(data)
print(df)

#then, put the following function to satisfy the given condition:
print(df[(df['A'] > 5) & (df['B'] < 10)])

# it will give those number wgich satisfyies the given condition
```

```python
Question 10: Given a pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that 
contains the moving average of the sales for past 7 days for eah row in the DataFrame. the moving average should be calculated using a window of size 7 
and should include the current day.
Answer:
Step 1: Importing Libraries

# importing Libraries
# importing pandas as pd
import pandas as pd

# importing numpy as np
# for mathematical calculations
import numpy as np

Step 2: Importing data
# importing time series data
reliance = pd.read_cvs('RELIANCE.NS.csv', index_col = 'Date', parse_dates = True)

# Printing dataframe
reliance.head()


Step 3: 
# updating our dataframe to have only
# one column 'Close' as rest all columns
# are of no use for us at the moment
# using .to_frame() to convert pandas series
# into dataframe.
reliance = reliance['close'].to_frame()

#calculating simple moving average
# using .rolling(windows).mean(),
# with window size = 7
reliance['SMA30'] = reliance['Close'}.rolling(7).mean()
                             
# removing all the NULL values using
# dropna () method
reliance.dropna(inplace = True)

# printing Dataframe
reliance
                             
```
